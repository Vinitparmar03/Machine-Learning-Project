```python
for col in df.select_dtypes(include='number').columns:
    mean = df[col].mean()
    std = df[col].std()
    upper = mean + 3 * std
    lower = mean - 3 * std
    outliers = df[(df[col] > upper) | (df[col] < lower)]
    df = df.drop(index=outliers.index)
    if not outliers.empty:
        print(f"\nColumn: {col}")
        print(f"Outlier count: {outliers.shape[0]}")
        print(outliers[[col]].head())
```

```python
import matplotlib.pyplot as plt

target = "price_range"
features = df.columns.drop(target)


plt.figure(figsize=(15, 4 * n_rows))

for i, col in enumerate(features):
    plt.subplot(n_rows, n_cols, i + 1)

    # If the feature is categorical (few unique values)
    if df[col].nunique() < 10:
        # Plot count of feature grouped by target
        cross = df.groupby([col, target]).size().unstack(fill_value=0)
        cross.plot(kind='bar', stacked=True, ax=plt.gca())
        plt.title(f"{col} distribution by {target}")
        plt.xlabel(col)
        plt.ylabel("Count")
    else:
        # Use boxplot: target classes vs numerical feature
        df.boxplot(column=col, by=target, ax=plt.gca())
        plt.title(f"{col} distribution by {target}")
        plt.suptitle("")  # Remove automatic title

    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
if "px_height" in df.columns and "px_width" in df.columns:
	df["screen_resolution"] = df["px_height"] * df["px_width"]
	df = df.drop(columns=["px_width", "px_height"])
df.head()
```

```python
if "sc_h" in df.columns and "sc_w" in df.columns:
	df["screen_size"] = df["sc_h"] * df["sc_w"]
	df = df.drop(columns=["sc_h", "sc_w"])

df.head()
```

```python
from sklearn.model_selection import train_test_split

X=df.drop("price_range", axis = 1)
y=df["price_range"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, stratify=y, random_state=42)
print(df.columns)
```

```python
X_train = pd.DataFrame(X_train, columns=X.columns)
y_train = pd.DataFrame(y_train, columns=["price_range"])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

scale_cols = [
  'battery_power',
  'fc',
  'int_memory',
  'm_dep',
  'mobile_wt',
  'n_cores',
  'pc',
  'ram',
  'talk_time',
  'clock_speed_log',
  'screen_resolution',
  'screen_size'
]

ct = ColumnTransformer([
    ("scaling", StandardScaler(), scale_cols)
])

X = ct.fit_transform(X_train)
```

```python
print(y_train.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score

models = {
    "Logistic Regression": LogisticRegression(solver='lbfgs', max_iter=1000),
    "Support Vector Machine (Linear)": LinearSVC(max_iter=10000),
    "Support Vector Machine (RBF)": SVC(kernel='rbf'),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB()
}

for name, model in models.items():
    score = cross_val_score(model, X_train, y_train.values.ravel(), cv=5, scoring="accuracy")
    print(f"{name}: Mean Accuracy = {score.mean():.4f}")
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
import seaborn as sns
import matplotlib.pyplot as plt

y_train = y_train.values.ravel()
y_test = y_test.values.ravel()

models = {
    "Random Forest": RandomForestClassifier(),
    "Naive Bays": GaussianNB(),
    "Decision Tree": DecisionTreeClassifier()
}

for name, model in models.items():
    model.fit(X_train, y_train)

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    train_acc = accuracy_score(y_train, y_train_pred)
    test_acc = accuracy_score(y_test, y_test_pred)

    precision = precision_score(y_test, y_test_pred, average='macro')
    recall = recall_score(y_test, y_test_pred, average='macro')
    f1 = f1_score(y_test, y_test_pred, average='macro')

    print(f"\n{name}")
    print(f"Train Accuracy: {train_acc:.4f}")
    print(f"Test Accuracy:  {test_acc:.4f}")
    print(f"Precision:      {precision:.4f}")
    print(f"Recall:         {recall:.4f}")
    print(f"F1 Score:       {f1:.4f}")

    print("\nClassification Report (Test):")
    print(classification_report(y_test, y_test_pred, digits=4))

    cm = confusion_matrix(y_test, y_test_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt']
}

rf = RandomForestClassifier(random_state=42)

grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    n_jobs=1,     
    scoring='accuracy',
    verbose=2
)

grid_search.fit(X_train, y_train)

print("\n✅ Best Parameters:")
print(grid_search.best_params_)

print("\n🏆 Best Cross-Validated Accuracy:")
print(grid_search.best_score_)

best_rf = grid_search.best_estimator_
```

```python
y_pred = best_rf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print("📊 Random Forest Evaluation on Test Set")
print(f"Accuracy  : {accuracy:.4f}")
print(f"Precision : {precision:.4f}")
print(f"Recall    : {recall:.4f}")
print(f"F1 Score  : {f1:.4f}\n")

print("📄 Classification Report:")
print(classification_report(y_test, y_pred, digits=4))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, cmap='Blues')
plt.title("Random Forest - Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```
